#### concatenate samps sequenced > 1 (or symlink) into fastqdir for dDocent

````python
# make sure all of the fastq data exists
counts = Counter()
for fq in df['fastq']:
    counts[op.exists(fq)] += 1
counts
````

````python
# are samples sequenced > 1. A: YES, therefore I need to concat files
luni(df['sample_name']), len(df.index)
````

````python
# convert samp names into something dDocent expects, will make it easier later
df['dDocent_sampname'] = [f'{pop}_{samp}' for (pop,samp) in zip(df['popn'], df['ind'])]
df.head()
````

````python
# make sure each sampname only has one "_"
for samp in df['dDocent_sampname']:
    assert len(samp.split("_")) == 2
````

````python
# get files for each sample (len(list) = count)
sampfiles = {}
for row in df.index:
    samp = df.loc[row, 'dDocent_sampname']
    if samp not in sampfiles:
        sampfiles[samp] = []
    sampfiles[samp].append(df.loc[row, 'fastq'])
````

````python
# create commands to concatenate samps sequenced > 1, otherwise just use a symlink to save space
def getdst(samp, num):
    # if symlink, it says unzipped (.fq suffix), but the src file is gzipped
    if num == 1:
        return op.join(fqdir, f'{samp}.F.fq.gz')
    else:
        return op.join(fqdir, f'{samp}.F.fq')

catdir = makedir(op.join(DIR, 'catfiles'))
commandfile = op.join(catdir, 'concatenate_fastqs.txt')
cmds = []
dsts = []
for samp,lst in sampfiles.items():
    if len(lst) == 1:
        dst = getdst(samp, 1)
        try:
            os.symlink(lst[0], dst)
        except:
            print("couldn't symlink %s" % src)
    else:
        dst = getdst(samp, 0)
        dsts.append(dst)
        cmds.append('zcat %s > %s' % (' '.join(lst), dst))
with open(commandfile, 'w') as o:
    o.write("%s" % '\n'.join(cmds))

# go gzip the concatenated files
gzipfile = op.join(catdir, 'gzip_concatenate_fastqs.txt')
cmds = []
for dst in dsts:
    cmds.append(f'gzip -f {dst}')
with open(gzipfile, 'w') as o:
    o.write("%s" % '\n'.join(cmds))
````
